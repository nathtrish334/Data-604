1.  What is a random seed in simulation, and why should we care?  How was simulation used in the article this week?

A random seed is used to initialize a random number generator. This is the value that a random number generator starts with when generating random numbers. This ensures that results are reproducible.

The Seed function saves the state of a random function, so that it can generate same random numbers on multiple executions of the code on the same machine or on different machines (for a specific seed value). The seed value is the previous value number generated by the generator. For the first time when there is no previous value, it uses current system time.

Common cases where random seeds are used include:
*   Model training in algorithms such as random forest and gradient boosting
*   Splitting data into training/validation/test sets. Random seeds ensure that the data is divided the same way every time the code is run
*    Bench-marking results of multiple versions of an algorithm. it’s important that all versions use the same data and are as similar as possible.
